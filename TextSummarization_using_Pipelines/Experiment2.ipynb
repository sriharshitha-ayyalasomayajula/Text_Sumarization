The goal here is to implement pre-trained text summarization pipelines from Hugging face and compare the output summaries. 

#####First install all the required libraries. First we'd like to understand summaires which are formed by extarcting important sentences from the input so we'll be applying LexRank and LSA summarizers. So, here we used sumy library to implement lexRank an LSA summarizers.

```python
!pip install sumy
import sumy
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.tokenizers import Tokenizer
from sumy.parsers.plaintext import PlaintextParser
from pprint import pprint
import nltk
nltk.download('punkt')
```

Here we have used Turning Red movie plot as the input text.

```python
#sample text to use for text extraction. The plot of movie Turning Red

input_text = """
In 2002 Toronto, 13-year-old Meilin "Mei" Lee, who lives with her parents, Ming and Jin, helps take care of the family's temple dedicated to their ancestor Sun Yee, and works to make her mother proud. She hides her personal interests from Ming, such as the fact that she and her friends Miriam, Priya, and Abby are fans of the boy band 4*Town. One night when Ming, who is strict and overprotective, discovers Mei's crush on Devon, the local convenience store clerk, she inadvertently humiliates Mei in public.

That night, Mei has a vivid nightmare involving red pandas. When she wakes up the next morning, she has transformed into a large red panda. She hides from her parents and discovers that she transforms only when she is in a state of high emotion. When she reverts to human form her hair remains red, and so she goes to school in a touque. Ming initially believes Mei is experiencing her first period, but learns the truth when she humiliates her at school, causing Mei to transform and run home in panic and tears.

Ming and Jin explain that Sun Yee was granted this transformation to protect her daughters and her village during wartime, and that all her female descendants have also had this ability. This has become inconvenient and dangerous in modern times, so the red panda spirit must be sealed in a talisman by a ritual on the night of a lunar eclipse, which will take place in a month's time. Mei's friends discover her transformation, but take a liking to it; Mei finds that concentrating on them enables her to control her transformations.

Ming allows Mei to resume her normal life, but refuses to let Mei attend 4*Town's upcoming concert. Instead, the girls secretly raise money for the tickets at school by exploiting the popularity of Mei's red panda form while lying to Ming about how Mei is spending her time. To raise the last $100, Mei agrees to attend school bully Tyler's birthday party as the red panda. At the party, Mei is upset to discover that the concert will be on the night she is to undergo the ritual. In her rage, she attacks Tyler when he insults her family, frightening the other kids. Ming discovers Mei's activities and accuses her friends of corrupting and taking advantage of her. Ashamed of her actions and afraid to stand up to her mother, Mei fails to come to her friends' defense.

To Ming's dismay, Mei's grandmother and aunts arrive to assist with Mei's ritual. As Mei prepares herself, Jin finds videos she recorded of herself as the red panda with her friends and tells her she should not be ashamed of this side of her, but to embrace it. During the ritual, as Mei's red panda form is about to be sealed, she decides to keep her powers and abandons the ritual to attend the concert at the SkyDome; in making her escape, she breaks Ming's talisman, releasing her red panda form as well. At the concert she reconciles with her friends and discovers that Tyler is also a 4*Town fan. However, an enraged Ming, having become a kaiju-sized red panda, disrupts the concert, intending to take Mei back by force.

Mei and Ming argue about the former's independence. As they quarrel, Mei accidentally knocks her mother unconscious. Mei's grandmother and aunts break their talismans to use their red panda forms to help drag Ming into a new ritual circle. Mei's friends and 4*Town join in singing to complete the ritual, sending Mei, Ming, and the other women to the astral plane. Mei reconciles with her mother. She helps her mend her bond with her own mother, whom Ming accidentally scarred in anger before her own red panda form was sealed. The other women contain their red pandas in new talismans; but Mei decides to keep hers, and Ming accepts that she is finding her own path.

Later, as the Lee family raises money to repair the damage to the SkyDome, Mei and Ming's relationship has improved. Mei balances her temple duties—where her red panda is now an attraction—with spending time with her friends, who now include Tyler." """

#pprint(input_sample_data)
```

```python
#parsing input text with the tokenizer 
string_parsing = PlaintextParser.from_string(input_text, Tokenizer("english"))
```

```python
#Applying LexRank Summarizer
#This top 10 best sentences by uses tf-idf metric (top 10 rankings), increases or decreases the number of sentences 
Lex = LexRankSummarizer()
string_lex_summary = Lex(document = string_parsing.document, sentences_count=10)
pprint(string_lex_summary)
```

```python
input_text = input_text.replace(u'\xa0', u' ')
len(input_text.split("\n"))
#Applying LSA Summarizer
#selects top 10 best sentences this time by LSA summarizer 
Lsa = LsaSummarizer()
string_lsa_summary = Lsa(document=string_parsing.document, sentences_count=10)
pprint(string_lsa_summary)
```

Here, we applied LexRank and LSA summarizers to the input data and the output sumamries are a combination of important sentences of the input text. We found LSA based summary to be a bit better than LexRank as it gives us a context and full summary of the input text. Yet we find both of them to be a bit vague and incomplete.   

So, we'd like to apply abstarctive text summarization methods to see if we can get a meaningful summary of the input text. For that we used transformer library.

Text summarization with Hugging face pipelines 

```python
#Installing transfromers 
!pip install transformers
! pip install transformers[sentencepiece]
from transformers import pipeline
```

```python
len(input_text.split())
#Model1 : Implementing the default summarizer - distilbart-cnn
summarizer = pipeline("summarization")
model1 = summarizer(input_text)
```

```python
psum1 = str(model1)
print("Default model Summary:")
print(psum1)
```

The default models yields a biref summary but it is incomplete, it doesnt give the complete summary.

Let's apply another model. 

```python
#Model2: Implementing pegasus-xsum model
pegasus_summarizer = pipeline("summarization", model="google/pegasus-xsum", tokenizer="google/pegasus-xsum", framework="pt")
model2 = pegasus_summarizer(input_text, min_length=50, max_length=150, truncation=True)
```

```python
psum2 = str(model2)
print("Pegasus model Summary:")
print(psum2)
```

The pegasus model also gives us a brief jist of the input text. This is better compared to the default model. 

```python
#Model3: Implementing the bart model
bart_summarizer = pipeline("summarization", model="facebook/bart-large", tokenizer="facebook/bart-large", framework="tf")
model3 = bart_summarizer(input_text, min_length =50, max_length = 100, truncation=True)
```

```python
psum3 = str(model3)
print("Bart model Summary:")
print(psum3)
```

We applied bart-large, this doesnt generate a meaningful summary. 
Let us apply mbart next.


```python
#Model4 : Implementing the mbart model
mbart_summarizer = pipeline ("summarization", model ="facebook/mbart-large-50", tokenizer = "facebook/mbart-large-50",use_fast = False)
model4 = mbart_summarizer(input_text, min_length = 50, max_length = 100, truncation = True)
```

```python
psum4 = str(model4)
print("Mbart model Summary:")
print(psum4)
```

Even this model doesnt generate a meaningful summary. 

We compare the generated summaries with the film synopsis.

```python
film_synopsis =""" A thirteen-year-old girl is torn between staying her mother's dutiful daughter and the changes of adolescence. And as if the challenges were not enough, whenever she gets overly excited she transforms into a giant red panda."""
print("Film Synopsis from the web:")
pprint(film_synopsis)
```

Evaluating the summaries with ROUGE scores

```python
!pip install rouge-score
from rouge_score import rouge_scorer
```

```python
#Rouge Score of model 1 - default model
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
model1_results = scorer.score(psum1, film_synopsis)
print("Default model Scores:")
pprint(model1_results)
```

```python
#Rouge Score of model 2 - pegasus
model2_results = scorer.score(psum2, film_synopsis)
print("Pegasus model Scores:")
pprint(model2_results)
```

```python
#Rougue Score of model 3 - bart
model3_results = scorer.score(psum3, film_synopsis)
print("Bart model Scores:")
pprint(model3_results)
```

```python
#Rouge socre of model 4 - mbart
model4_results = scorer.score(psum4, film_synopsis)
print("Mbart model Scores:")
pprint(model4_results)
```
